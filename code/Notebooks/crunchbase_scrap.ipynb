```python
df1 = pd.read_csv(
    "/Users/cordiez/Documents/CompaniesHouse/OutputFiles/CB_2.csv", index_col=0
)


ch = ""
for elt in df1["Industries"].tolist():
    try:
        ch = ch + elt
        ch = ch + ","
    except:
        pass

codes = list(set(ch.split(",")))
codes_cleaned = []
for elt in codes:
    if len(elt) > 0:
        if elt[0] == " ":
            x = elt[1:]
            codes_cleaned.append(x)
        else:
            x = elt
            codes_cleaned.append(x)
len(codes_cleaned)
```

```python
df = pd.DataFrame(codes_cleaned)
df.to_csv(
    "/Users/cordiez/Documents/CompaniesHouse/OutputFiles/crunchbase_industries.csv"
)
```
